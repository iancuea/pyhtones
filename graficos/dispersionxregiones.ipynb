```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")

df = pd.read_csv("csv/ArchivoC_Adm2026REG.csv", sep=";")

df["CLEC_REG_ACTUAL"] = pd.to_numeric(df["CLEC_REG_ACTUAL"], errors="coerce")
df["MATE1_REG_ACTUAL"] = pd.to_numeric(df["MATE1_REG_ACTUAL"], errors="coerce")

regiones = {
    1: "Tarapacá",
    2: "Antofagasta",
    3: "Atacama",
    4: "Coquimbo",
    5: "Valparaíso",
    6: "O'Higgins",
    7: "Maule",
    8: "Biobío",
    9: "Araucanía",
    10: "Los Lagos",
    11: "Aysén",
    12: "Magallanes",
    13: "Metropolitana",
    14: "Los Ríos",
    15: "Arica y Parinacota",
    16: "Ñuble"
}

df["REGION"] = df["CODIGO_REGION"].map(regiones)
```

```python
datos = df[
    (df["CLEC_REG_ACTUAL"].notna()) &
    (df["CLEC_REG_ACTUAL"] > 0)
]

media = datos["CLEC_REG_ACTUAL"].mean()
moda = datos["CLEC_REG_ACTUAL"].mode()[0]
desv = datos["CLEC_REG_ACTUAL"].std()

plt.figure(figsize=(18,8))

sns.stripplot(
    data=datos,
    x="REGION",
    y="CLEC_REG_ACTUAL",
    jitter=0.28,
    alpha=0.35,
    size=3
)

plt.axhline(media, color="red", linestyle="--", linewidth=2,
            label=f"Media {media:.1f}")

plt.axhline(moda, color="purple", linestyle="-.", linewidth=2,
            label=f"Moda {moda:.1f}")

plt.axhline(media + desv, color="green", linestyle=":",
            linewidth=2, label="+1 Desv. Std")

plt.axhline(media - desv, color="green", linestyle=":",
            linewidth=2, label="-1 Desv. Std")

plt.title("Competencia Lectora PAES por Región", fontsize=16)
plt.xlabel("Región")
plt.ylabel("Puntaje")
plt.xticks(rotation=75, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```
